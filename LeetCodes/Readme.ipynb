3. 背诵指数 *****   
注意：理论上应该存储m x n 尺寸的表格以记录各个点对应的最大路径数目，并由此计算动态方程$dp[i][j] = dp[i-1][j] + dp[i][j-1]$。但是存储一整个表格的复杂度是$O(mxn)$，而实际上计算动态方程需要的只是当前元素紧邻上一行和上一列的表格数值。这种情形一定可以简化成仅存储两行（before、curr），然后通过$curr[j]=before[j]+curr[j-1]$来更新。进一步，before可以省略为curr向量上一次循环结束后的值，从而实现$O(n)$的空间复杂度
